```python
movies=movies[['id','imdb_id','budget','overview']]
movies
```

```python
new_data=movies.drop(columns=['overview','budget'])
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000,stop_words='english')
cv
```

```python
vector=cv.fit_transform(new_data['id'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['id']==168259].index[0]
```

```python
distance=sorted(list(enumerate(similarity[4])),reverse=True,key=lambda vector:vector[1])
```

```python
for i in distance[0:5]:
  print(new_data.iloc[i[0]].id)
```

```python
def recommand(movies):
  index=new_data[new_data['id']==movies].index[0]
  distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
  for i in distance[0:5]:
    print(new_data.iloc[i[0]].id)
```

```python
recommand(262500)
```
